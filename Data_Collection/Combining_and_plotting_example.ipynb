### Reads the data file and prints the totals for each user

```python
df = pd.read_csv('spotify_tracks_LiveCode.csv')
df['user_id'].value_counts()
```

### Creates a new table from the df object above

```python
df_audio_features_mean = pd.DataFrame()

for user_id in df['user_id'].unique():
    df_audio_features_temp = df.query(f"user_id=='{user_id}'").loc[:, 'danceability':'valence']
    df_audio_features_temp = df_audio_features_temp.drop(['key','mode','loudness'], axis = 1)
    df_audio_features_temp = df_audio_features_temp.mean().reset_index()
    df_audio_features_temp.columns = ['audio_feature','mean']
    df_audio_features_temp['user_id'] = user_id
    
    df_audio_features_mean = pd.concat([df_audio_features_mean, df_audio_features_temp.copy()])
    
    del df_audio_features_temp

df_audio_features_mean = df_audio_features_mean.reset_index(drop = True)
```

### Plots the filtered data as a radar plot

```python
import plotly.express as px

fig = px.line_polar(df_audio_features_mean, r='mean', theta='audio_feature',
                    color = 'user_id', line_close=True)
# fig.update_traces(fill='toself')
fig.show()
```

### Plotting 3D of all songs

```python
fig3d = px.scatter_3d(df, x='acousticness', y='instrumentalness',
                      z='liveness', color='user_id')
fig3d.show()
```

### Plots a bar chart for each user's tracks by release date.

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_year'] = [date.year for date in df['release_date']]
df_year_counts_by_user = df[['user_id','release_year']].groupby('user_id').value_counts().reset_index().rename(columns={0:'count'})

figbar = px.bar(df_year_counts_by_user, x="release_year", y="count", color="user_id", title="Tracks by Time")

figbar.show()
```
